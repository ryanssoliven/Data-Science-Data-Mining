```python
# Decision Tree Implementation using Gini

from sklearn.metrics import accuracy_score

# Library to load data in dataframes
import pandas as pd

# Library for arrays
import numpy as np

# Library for DecisionTree
from sklearn.tree import DecisionTreeClassifier, export_graphviz

# Library to create DecisionTree Graph
import pydotplus
from IPython.display import Image
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import tree

# Load the data in the pandas dataframe
dp = pd.read_csv("heart-disease-dataset2.csv")
dp = dp.replace('?', 0)
dp = dp.dropna()

# Printing the dataset shape
print("Dataset Length: ", len(dp))
print("Dataset Shape: ", dp.shape)

# Reencode data
X = pd.get_dummies(dp[['age', 'sex', 'cp', 'tresbps', 'chol', 'fbs', 'restecg', 'thalach','exang', 'oldpeak', 'slope', 'ca', 'thal']].astype('float'))
#X = pd.get_dummies(dp[['sex', 'cp', 'fbs', 'restecg', 'slope', 'thal']].astype('float'))

# Get the target column - result
Y = dp.values[:,13].astype('float')

X_train_1, X_test_1, Y_train_1, Y_test_1 = train_test_split(X, Y, test_size=0.2, random_state=5)

# DecisionTree using Gini Index
ds_gini = DecisionTreeClassifier(criterion='gini')

# Performing training
ds_train = ds_gini.fit(X_train_1, Y_train_1)
# Testing tree
y_pred = ds_train.predict(X_test_1)

print("Gini Model Accuracy Score: {}%".format(accuracy_score(Y_test_1, y_pred)*100))

# Create Dot data
dot_data = export_graphviz(ds_train, out_file=None, feature_names=list(X_train_1.columns.values), class_names=['0','1','2','3','4'], rounded=True, filled=True)

# Create Graph from Dot data
graph = pydotplus.graph_from_dot_data(dot_data)

print(graph)

# Show graph
Image(graph.create_png())
```
