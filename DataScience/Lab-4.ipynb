```python
# Library to load data in dataframes
import pandas as pd

# Library for arrays
import numpy as np

# Library for regression analysis
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score

# Data visualization library
from matplotlib import pyplot as plt
import seaborn as sns

# Load in the data into pandas
data = pd.read_csv("student_score.csv")

# Load the data in the pandas dataframe
dp = pd.DataFrame(data, columns=['Student','Xi','Yi'])

# Take X_ts as X variable for simple linear regression
# Take y_sg as Y variable for simple linear regression
X_ts = dp.Xi
y_sg = dp.Yi

# Normalize the X and Y between -1,1
X_ts = np.array(X_ts).reshape(-1,1)
y_sg = np.array(y_sg).reshape(-1,1)

# Take linear regression
lr = linear_model.LinearRegression()

# Fit the data
lr.fit(X_ts,y_sg)

# Plot the regression graph (Exercise 2)
prediction_space = np.linspace(min(X_ts), max(X_ts)).reshape(-1,1)
plt.scatter(X_ts, y_sg)
plt.plot(prediction_space, lr.predict(prediction_space), color='red', linewidth=3)
plt.ylabel("Statistics Grade")
plt.xlabel("Test Score")
plt.show()

# Measure the performance R-squared (Exercise 3)
r = round(lr.score(X_ts, y_sg), 2)
print("The model performance")
print("-----------------------------")
print("R-squared score is {}".format(r))
print("\n")
```

```python
# Library to load data in dataframes
import pandas as pd

# Library for arrays
import numpy as np

# Library for regression analysis
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score

# Data visualization library
from matplotlib import pyplot as plt
import seaborn as sns

# Load in the data into pandas
diabetes = pd.read_csv("Dataset heart-disease-dataset.csv")

# Logistic Regression
X_age = diabetes['age']
y_thalach = diabetes['thalach']

# Let's normalize the X and Y between -1 and 1
X_age = np.array(X_age).reshape(-1,1)
y_thalach = np.array(y_thalach).reshape(-1,1)

# Build the regression model
# Split the dataset 80% for training and 20% for validation
X_train_1, X_test_1, Y_train_1, Y_test_1 = train_test_split(X_age, y_thalach, test_size=0.2, random_state=5)

# Take non-linear regression
lr = linear_model.LogisticRegression(max_iter=10000)

# Fit the data
lr.fit(X_train_1, Y_train_1)

# Get prediction for the test data
y_pred_1 = lr.predict(X_test_1)

# Measure the performance MSE.R-squared
mse = mean_squared_error(Y_test_1, y_pred_1)
r = round(lr.score(X_test_1, Y_test_1), 2)

print("The model performance")
print("-----------------------------")
print("MSE is {}".format(mse))
print("R-squared score is {}".format(r))
print("\n")

# Plot the regression graph
prediction_space = np.linspace(min(X_age), max(X_age)).reshape(-1,1)
plt.scatter(X_age, y_thalach)
plt.plot(prediction_space, lr.predict(prediction_space), color='red', linewidth=3)
plt.ylabel("Thalach")
plt.xlabel("Age")
plt.show()
```

According to my results, the simple linear regression model is more accurate for prediction than the logistic regression model because:

1) The MSE for the simple linear regression model < than the MSE for logistic regression (356.9509374452787 < 413.42622950819674)

2) The R-squared value for the simpler linear regression model > than the MSE for logistic regression (0.2 > 0.02)
